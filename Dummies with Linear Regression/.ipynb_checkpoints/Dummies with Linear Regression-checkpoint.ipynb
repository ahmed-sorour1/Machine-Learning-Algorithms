```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
conc = pd.concat([df, dummies],axis = 'columns')
final_df = conc.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
final_df
```

### Audi A5 == > 1,0 
### BMW  X5 == > 0,1 
### Mercedez == > 0,0

```python
model = LinearRegression()
x = final_df.drop('Sell Price($)', axis= 'columns')
y = final_df['Sell Price($)']
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
#Predict price of a mercedez benz that is 4 yr old with mileage 45000
print(int(model.predict([[45000,4,0,0]])))
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86000
print(int(model.predict([[86000,7,0,1]])))
```
